# Accuracy

```python
feature_selection_schemes = ['Base Classifier', 'Pearson Correlation', 'Hypothesis Testing', 
                            'Decision Tree', 'Logistic Regression', 'Support Vector']

datasets = ['CIC-IDS-2017', 'UNSW-NB15', 'NF-UNSW-NB15-v2', 'CSE-CIC-IDS2018', 'NSL-KDD']

test_accuracy = [
       [0.9978, 0.9756, 0.9756, 0.9981, 0.9971, 0.9759],
       [0.9992, 0.9994, 0.9994, 0.9915, 0.9976, 0.9988],
       [0.9974, 0.9972, 0.9972, 0.9954, 0.9968, 0.9969],
       [0.9999, 0.9985, 0.9985, 0.9996, 0.9986, 0.9987],
       [0.9920, 0.9878, 0.9881, 0.9600, 0.9729, 0.9737]
]

colors = ['#003f5c', '#58508d', '#bc5090', '#ff6361', '#ffa600']
```

```python
for i in range(5):
    x = feature_selection_schemes
    y = test_accuracy[i]
    plt.plot(x, y, color=colors[i], label=datasets[i], marker='o', alpha=1)

plt.ylim(0.955, 1.005)
plt.xticks(rotation=60)
plt.legend()
plt.savefig('plots/accuracyResults.png', dpi=300, bbox_inches='tight')
```
